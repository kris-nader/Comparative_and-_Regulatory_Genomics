Cannadidiol is a phytocannabinoid that is derived from the hemp plant.  CBD products have risen in popularity with applications ranging from treatment for skin conditions, potential uses in treatment of cancer and as palliative care. This research aims to 
characterize the effects of CBD on human keratinocytes using bulk-RNA sequencing and sequential window acquistiion of all theoretical mass spectrometry. The data in this paper consists of 3 replicates each: cells were either cultured in DMSO(Control- 3 replicates ) or cultured in CBD(Experiemntal -3 replicates). For the work presented today, only 2 repicates from each experiemtal condition is taken. Genes that were of specific interest were BACH1, a gene encoding for a transcriptional regulator protein BTB and CNC homology 1 and HMOX1, heme oxygenase 1 which make up parts of the BACH1/HO-1 pathway. HMOX1/HO-1 catalyzes the degradation of proteins containing the heme group and is induced in times of cellular stress. NRF2, a nuclear factor erythroid 2-like 2 gene, encoding a transcriptonal activator of HMOX1 will also be an important element in downstream analysis. 

In order to help me understand more about the general pathways, I found an image in a Review paper that I will briefly explain:

In an situation without oxidative stress, NRF2 is bound by KEAP1 which faciliates its degradation in the cytoplasm. During oxidative stress, NRF2 dissociates from Keap1 and accumulates in the cytoplasm where it is then transported into the nucleus. Nrf2/NFE2L2, now in the nucleus, binds to MAREs(Maf recognition elements) with small protein Maf therefore activating HMOX1, a gene important in oxidative stress response. BACH1 competes with Nrf for binding to MAREs. Therefore, BACH1 represents the transcriptional repressor while NRF2 represents the transcriptional activator of HMOX1/HO-1.

HMOX1 is upregulated in the Experimental data. It is induced by NRF2 which also activates other genes such as SQSTM1 /p62 and GCLC. The paper revealed that CBD is a BACH1 inhibitor and weak NRF2 activator. I originally assumed that BACH1 would be down-regulated.I had come to this conclusion upon looking at barplots for the gene counts in the figure below and the logFC of the genes. However, this is not the case. BACH1 revealed to be upregulated in the experimental setup. 


However, upon reading more articles on the BACH1-HMOX1 pathways I discovered that CBD(experimental) was involved in the nuclear export and degradation of BACH1, a transcription repressor of HMOX1( an important detail)! Therefore, by exporting BACH1 out of the nucleus, it is unable to repress HMOX1 expression. Therefore, BACH1 is not down-regulated, but targeted and finding a significant p-value and positive logFC is logical. We would also expect other target genes of NRF2 to be upregulated in the experimental data: GCLC and SQSTM1/p62.

The barplots above give important insight into the biological process and validate initial findings. NRF2/NFE2L2 is up-regulated in the experimental data. This means there is more expression as compared to the control and will then be able to upregulate target genes. These target genes as mentioned before are GCLC, SQSTM1 and HMOX1. The plots show this upregulated of the target genes in an easy visualization step. The logFC can also be used to check for up-regulation but using a visual aid helped in quickly grasping the concept. The graphs, plots and queries for the logFC were derived using the 2nd notebook for differential gene expression: kristen_DeSEQ.ipynb.
